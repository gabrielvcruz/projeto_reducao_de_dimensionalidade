```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# 1️⃣ Carregar o modelo YOLOv8 para detecção de rostos
model = YOLO("yolov8n.pt")  # Utilizando o modelo YOLOv8 genérico

# 2️⃣ Carregar imagem
img_path = "jogadores_selecao.jpg"  # Substitua pelo caminho da sua imagem
img = cv2.imread(img_path)

# 3️⃣ Realizar a detecção
results = model(img)

# 4️⃣ Mostrar as caixas delimitadoras das faces detectadas
for r in results:
    for box in r.boxes.xyxy:
        x1, y1, x2, y2 = map(int, box)  # Coordenadas da bounding box
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Desenhar a caixa

# 5️⃣ Exibir imagem com as detecções
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
