```python


# Clonar o repositório do Darknet
!git clone https://github.com/AlexeyAB/darknet.git
%cd darknet
!make

# Baixar pesos pré-treinados do YOLOv4
!wget https://pjreddie.com/media/files/yolov4.weights

# Baixar e preparar o COCO Dataset
!mkdir /content/datasets
%cd /content/datasets
!wget http://images.cocodataset.org/zips/train2017.zip
!wget http://images.cocodataset.org/annotations/annotations_trainval2017.zip
!unzip train2017.zip
!unzip annotations_trainval2017.zip

# Criar arquivo de configuração (coco.data)
with open('/content/darknet/cfg/coco.data', 'w') as f:
    f.write("""classes=80
train="/content/datasets/train2017"
valid="/content/datasets/val2017"
names="/content/darknet/cfg/coco.names"
backup="/content/darknet/backup/"
""")

# Iniciar o treinamento
%cd /content/darknet
!./darknet detector train cfg/coco.data cfg/yolov4.cfg yolov4.weights -dont_show -map

# Testar o modelo treinado com uma imagem
!./darknet detector test cfg/coco.data cfg/yolov4.cfg backup/yolov4_final.weights /content/datasets/train2017/000000000001.jpg

# Exibir o resultado da inferência
from IPython.display import display
from PIL import Image
image = Image.open("predictions.jpg")
display(image)

```
